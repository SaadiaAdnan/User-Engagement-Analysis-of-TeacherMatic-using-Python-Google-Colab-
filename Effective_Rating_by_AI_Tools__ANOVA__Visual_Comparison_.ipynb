```python
# Install required packages
!pip install pandas scipy seaborn matplotlib

# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# Sample data (simulated for demonstration):
# Assume effectiveness ratings (1–5) for each AI tool by user group

# Create sample data: each list represents user ratings for a tool
chatgpt_ratings = [4, 4, 5, 3, 4, 4, 5, 3, 4, 3, 4, 5, 3, 4, 3]
teachermatic_ratings = [3, 3, 2, 4, 3, 2, 3, 2, 3, 4]
copilot_ratings = [2, 3, 3, 3, 3, 2, 3]

# Run one-way ANOVA
f_stat, p_value = f_oneway(chatgpt_ratings, teachermatic_ratings, copilot_ratings)

# Display results
print(f"F-statistic: {f_stat:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("✅ Significant difference in effectiveness ratings between AI tools.")
else:
    print("❌ No significant difference in effectiveness ratings between AI tools.")

# Optional: Create boxplot for visual comparison
ratings_df = pd.DataFrame({
    'Rating': chatgpt_ratings + teachermatic_ratings + copilot_ratings,
    'Tool': ['ChatGPT'] * len(chatgpt_ratings) +
            ['TeacherMatic'] * len(teachermatic_ratings) +
            ['Copilot'] * len(copilot_ratings)
})

plt.figure(figsize=(8, 5))
sns.boxplot(x='Tool', y='Rating', data=ratings_df)
plt.title("Effectiveness Ratings by AI Tool")
plt.ylim(1, 5)
plt.grid(True)
plt.show()
```
