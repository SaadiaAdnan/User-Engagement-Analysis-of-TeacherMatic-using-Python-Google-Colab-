```python
# Install required libraries (if not already installed)
!pip install pandas scipy matplotlib seaborn

# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Sample data: Cross-tabulation of AI Tools vs Frequency of Use
# These values are estimated based on the survey images provided
data = {
    'Daily':        [4, 1, 1, 0, 0],   # ChatGPT, TeacherMatic, Copilot, Other, None
    'Weekly':       [14, 6, 4, 2, 0],
    'Monthly':      [9, 5, 3, 2, 1],
    'Rarely':       [10, 5, 0, 3, 4],
    'Never':        [2, 1, 0, 1, 8]
}

# Row labels
tools = ['ChatGPT', 'TeacherMatic', 'Copilot', 'Other', 'None']

# Create DataFrame
df = pd.DataFrame(data, index=tools)

# Display data
print("Contingency Table (AI Tool vs Frequency of Use):")
display(df)

# Chi-Square Test of Independence
chi2, p, dof, expected = chi2_contingency(df)

# Output test results
print(f"\nChi-Square Statistic: {chi2:.2f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}")

if p < 0.05:
    print("✅ There is a statistically significant association between AI tool and frequency of use.")
else:
    print("❌ There is no statistically significant association between AI tool and frequency of use.")

# Visualize with a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, fmt="d", cmap="YlGnBu")
plt.title("AI Tool Usage by Frequency (Heatmap)")
plt.xlabel("Frequency of Use")
plt.ylabel("AI Tool")
plt.tight_layout()
plt.show()
```
